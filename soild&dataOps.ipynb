```python
Difference between ‘is-a’ relationship and ‘has-a’ relationship ?

##‘is – a’ relationship
1-The ‘is – a’ relationship is expressed with Inheritance ,
2-Class Inheritance,
3-Inheritance uses extend keyword
For example House is a Building. But Building is not House
Is a [House is a Building]
class Building
{
 ……….
}

class House extends Building
{
………..
}
##‘has – a’ relationship
1-The ‘has – a’ relationship is expressed with Composition.
2-Object Composition
3-Composition uses new keyword.
for example House has a Bathroom.
Has a [House has a Bathroom]
class Bathroom
{
 …………
}

class House
{
  Bathroom room=new Bathroom();
  ………..
}
```

```python
Introduction to Kubernetes?
Kubernetes is an open-source Container Management tool w
hich automates container deployment, 
container scaling, and descaling and container load balancing 
It is written in Golang and has a huge community because it was first developed by Google
and later donated to CNCF
Kubernetes can group ‘n’ number of containers into one logical unit for managing and deploying them easily.
It works brilliantly with all cloud vendors i.e. public, hybrid and on-premises
***********************************************************************************************
Kubernetes is an open-source platform that manages Docker containers in the form of a cluster.
Along with the automated deployment and scaling of containers,
it provides healing by automatically restarting failed containers and rescheduling them when their hosts die. 
This capability improves the application’s availability
```

```python
What Is DataOps?
DataOps is a collection of technical practices, workflows, cultural norms, and architectural patterns
1-Rapid innovation and experimentation delivering new insights to customers with increasing velocity
2-Extremely high data quality and very low error rates
3-Collaboration across complex arrays of people, technology, and environments
4-Clear measurement, monitoring, and transparency of results
```
